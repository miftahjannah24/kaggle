```python
import kagglehub
import pandas as pd
import os

# Download dataset langsung ke path sementara
path = kagglehub.dataset_download("imtkaggleteam/agriculture-dataset-karnataka")

# Cek path dataset
print("Path to dataset files:", path)

# Daftar semua file di dalam folder dataset
files = os.listdir(path)
print("Files in dataset:", files)

# Cari file CSV di dalam folder
csv_files = [f for f in files if f.endswith('.csv')]

if csv_files:
    # Ambil file pertama yang memiliki format CSV
    csv_path = os.path.join(path, csv_files[0])
    
    # Baca file CSV
    df = pd.read_csv(csv_path)
    print(df.head())  # Menampilkan 5 baris pertama
else:
    print("No CSV files found in the dataset.")
```

```python
# Mengecek jika ada missing value di DataFrame
print("Apakah ada NaN atau missing value di data:")
print(df.isna().any().any())  # Mengembalikan True jika ada NaN

# Mengecek jumlah missing value per kolom
print("\nJumlah missing value per kolom:")
print(df.isna().sum())
```

```python
# Pilih kolom yang diinginkan
df2 = df[['Year', 'Location', 'Area']]

# Tampilkan dataset baru
print("Dataset baru:")
print(df2)

# Menyimpan dataset baru ke file CSV
df2.to_csv('data_season2.csv', index=False)
```

```python
# Tampilkan nama-nama kolom
print("Label kolom pada dataset:")
print(df.columns)
```

```python
# Pilih baris dengan indeks ganjil
df3_ganjil = df2.iloc[1::2]  # Mulai dari indeks 1, lalu lompat setiap 2 baris

# Tampilkan dataset baru
print("Dataset baru dengan indeks ganjil:")
print(df3_ganjil)

# Menyimpan dataset baru ke file CSV
df3_ganjil.to_csv('data_season3.csv', index=False)
```

```python
# Pilih baris dengan indeks ganjil
df2_ganjil = df.iloc[1::2]  # Mulai dari indeks 1, lalu lompat setiap 2 baris

# Pilih kolom yang diinginkan
df4 = df[['Rainfall']]

# Pilih baris dengan indeks ganjil
df4_ganjil = df4.iloc[1::2]  # Mulai dari indeks 1, lalu lompat setiap 2 baris

# Baca dataset yang sudah ada
df_lama = df2_ganjil

# Baca dataset baru yang kolomnya akan ditambahkan
df_baru = df4_ganjil

# Gabungkan kedua dataset berdasarkan kolom
df_gabungan = pd.concat([df_lama, df_baru], axis=1)

# Tampilkan dataset gabungan
print("Dataset gabungan:")
print(df_gabungan)

# Menyimpan dataset gabungan ke file CSV
df_gabungan.to_csv('data_season4.csv', index=False)
```

```python
# Mengelompokkan data berdasarkan kolom "Crop_Type"
df_grouped = df_gabungan.groupby('Temperature')

# Menggabungkan semua grup menjadi satu DataFrame
df_all_groups = df_grouped.apply(lambda x: x)

# Reset index agar lebih rapi
df_all_groups.reset_index(drop=True, inplace=True)

# Menampilkan semua data
print(df_all_groups)
```

```python
# Menampilkan semua grup dengan iterasi manual
for group_name, group_data in df_grouped:
    print("="*40)
    print(f"Group: {group_name}")
    print(group_data)
```
